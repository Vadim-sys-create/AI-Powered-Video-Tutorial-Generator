```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Загрузим изображение
image_path = "/content/Траволта_безфона.mp4"  # Замени на путь к своему изображению
image = Image.open(image_path)
draw = ImageDraw.Draw(image)

# Настройки текста
text = "Good!"
font = ImageFont.load_default()  # Используем стандартный шрифт
text_color = (255, 255, 255)  # Белый цвет текста

# Параметры анимации
text_x = image.width  # Начальная позиция текста (за пределами изображения)
text_y = image.height // 2  # Позиция текста по вертикали
speed = 5  # Скорость движения текста

# Функция для обновления кадра анимации
def update(frame):
    global text_x
    draw.rectangle((0, 0, image.width, image.height), fill=(0, 0, 0))  # Очистка кадра
    draw.text((text_x, text_y), text, font=font, fill=text_color)  # Рисуем текст
    text_x -= speed  # Двигаем текст влево
    if text_x < -draw.textlength(text, font=font):  # Если текст ушел за пределы, возвращаем его
        text_x = image.width
    plt.imshow(image)
    plt.axis('off')

# Создаем анимацию
fig = plt.figure()
ani = FuncAnimation(fig, update, frames=100, interval=50)

# Показываем анимацию
plt.show()
```

```python
import cv2

# Загрузим видео
video_path = "/content/Траволта_безфона.mp4"
output_path = "/content/Траволта_с_текстом.mp4"

cap = cv2.VideoCapture(video_path)

# Получаем параметры видео
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Создаем объект для записи видео
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Настройки текста
text = "Где все?"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 2
font_color = (255, 255, 255)  # Белый цвет
thickness = 3

# Параметры анимации
text_x = width  # Начальная позиция текста (за пределами кадра)
text_y = height // 2  # Позиция текста по вертикали
speed = 10  # Скорость движения текста

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Добавляем текст на кадр
    cv2.putText(frame, text, (text_x, text_y), font, font_scale, font_color, thickness)
    text_x -= speed  # Двигаем текст влево
    if text_x < -cv2.getTextSize(text, font, font_scale, thickness)[0][0]:  # Если текст ушел за пределы, возвращаем его
        text_x = width

    # Записываем кадр в выходное видео
    out.write(frame)

# Освобождаем ресурсы
cap.release()
out.release()
print(f"Видео сохранено в {output_path}")
```
